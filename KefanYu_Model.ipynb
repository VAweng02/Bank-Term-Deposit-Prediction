```python
df['y_encode'] = df['y'].map({'no': 0, 'yes': 1})
df.drop("y", axis=1, inplace=True)
```

```python
X = df.iloc[:, 0:20]  # Set X be all columns except target
y = df.y_encode
```

```python
X_dur = X.drop('duration', axis=1)
```

```python
# split X and y into training and testing sets
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.2)  # have a 80-20 split for train and test of X and y
```

```python
print(X_train.shape)
print(X_test.shape)
print(y_train.shape)
print(y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression

reg = LogisticRegression()
reg.fit(X_train, y_train)  # fit the model
y_pred = reg.predict(X_test)  # make prediction
```

```python
from sklearn import metrics
import matplotlib.pyplot as plt
import seaborn as sns

cnf_matrix = metrics.confusion_matrix(y_test, y_pred)  # create the confusion matrix
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')  # show in the form of heat map

plt.title('Confusion Matrix', fontsize = 15)
plt.xlabel('Prediction', fontsize = 10)
plt.ylabel('Truth', fontsize = 10)

plt.show()
```

From the confusion matrix, we can see that the number of correctly predicted labels for y = no is 6950, and the number of correctly predicted labels for y = yes is 310. it is normal for such imbalanced distribution of yes and no values. With such values, we have the following

```python
from sklearn.metrics import (
    accuracy_score,
    precision_score,
    recall_score,
    f1_score
)
```

```python
accuracy = accuracy_score(y_test, y_pred)  # calculate accuracy score
precision = precision_score(y_test, y_pred)  # calculate precision score
recall = recall_score(y_test, y_pred)  # calculate recall score
f1score = f1_score(y_test, y_pred)  # calculate f1  score

print(f"Accuracy = {accuracy}")
print(f"Precision = {precision}")
print(f"Recall = {recall}")
print(f"F1 Score = {f1score}")
```

We can see that the accuracy score is around 92%, which is fairly good and 92% of labels are correctly predicted. The precision is about 64%, which indicates that 64% of positive identifications are accurate. In other words, then ratio of the number of samples correctly predicted as no and the number of samples predicted as no is 64%, which is not that high. The recall is around 41%, which indicates the ratio of the number of samples correctly predicted as yes and the number of samples predicted as yes is 41%. The reason for such low score is that the number of yes subscription is very low, and the type II error occurs more furequently due to such imbalance. The f1 score is the mean between recall and precision, which is around 50%

```python
y_pred_proba = reg.predict_proba(X_test)[::,1]
fpr, tpr, _ = metrics.roc_curve(y_test,  y_pred_proba)  # find false positive and true positive rate
auc = metrics.roc_auc_score(y_test, y_pred_proba)  # get auc
plt.plot(fpr,tpr,label="auc="+str(auc))
plt.legend(loc=4)
plt.show()
```

The AUC is close to 1 , indicating high value of correct predictions

In conclusion, Logistic Regression is a fairly good model with very high accuracy, but since the number of no is far more than yes, which leads to a relatively low precision and recall score

Since this dataset is imbalanced, Iwe try another approach to resample it

```python
from sklearn.model_selection import StratifiedKFold
from imblearn.over_sampling import SMOTE
from sklearn.linear_model import LogisticRegression
from sklearn import metrics

kfolds = StratifiedKFold(n_splits=5)
for ind,(ind_train,ind_val) in (enumerate (kfolds.split(X,y))):
    X_train,X_test = X.iloc[ind_train],X.iloc[ind_val] 
    y_train,y_test = y.iloc[ind_train],y.iloc[ind_val]
    sm = SMOTE(random_state=42)

    X_train_res, y_train_res = sm.fit_resample(X_train, y_train)
    
    reg = LogisticRegression()
    reg.fit(X_train_res, y_train_res)  # fit the model
    y_pred = reg.predict(X_test)  # make prediction
    
cnf_matrix = metrics.confusion_matrix(y_test, y_pred)
sns.heatmap(pd.DataFrame(cnf_matrix), annot=True, cmap="YlGnBu" ,fmt='g')  # show in the form of heat map

plt.title('Confusion Matrix', fontsize = 15)
plt.xlabel('Prediction', fontsize = 10)
plt.ylabel('Truth', fontsize = 10)

plt.show()
```
